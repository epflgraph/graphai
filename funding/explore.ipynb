```python
import pandas as pd
import matplotlib.pyplot as plt

from interfaces.db import DB
```

### Fetch data from db

```python
db = DB()

limit = 100000
```

```python
# Organisations
orgs = db.get_organisations(limit=limit)
org_cols = ['org_id', 'org_type', 'org_name', 'op_status', 'founding_date', 'termination_date', 'city', 'region', 'country', 'n_employees', 'last_funding_date', 'n_funding_rounds', 'funding_currency', 'total_funding', 'total_funding_usd', 'roles', 'rank']
orgs = pd.DataFrame(orgs, columns=org_cols)

orgs.head()
```

```python
# Funding rounds
frs = db.get_funding_rounds(limit=limit)
fr_cols = ['fr_id', 'fr_type', 'fr_name', 'fr_date', 'fr_currency', 'fr_amount', 'fr_amount_usd', 'fr_amount_chf', 'fr_amount_present_value_usd', 'fr_amount_present_value_chf', 'post_fr_valuation_cur', 'post_fr_valuation', 'post_fr_valuation_usd', 'city', 'region', 'country', 'n_investors', 'rank']
frs = pd.DataFrame(frs, columns=fr_cols)

frs.head()
```

```python
# People
people = db.get_people(limit=limit)
person_cols = ['person_id', 'sciper', 'last_name', 'first_name', 'full_name', 'gender']
people = pd.DataFrame(people, columns=person_cols)

people.head()
```

```python
# EPFL Startups
startups = db.get_startups(limit=limit)
startup_cols = ['startup_id', 'startup_name', 'legal_entity', 'founding_year', 'exit_year', 'status', 'industry']
startups = pd.DataFrame(startups, columns=startup_cols)

startups.head()
```

```python
# Concepts
concepts = db.get_concepts(limit=limit)
concept_cols = ['concept_id', 'concept_name']
concepts = pd.DataFrame(concepts, columns=concept_cols)

concepts.head()
```

```python
# Association: Funding rounds <-> Organisations
frs_orgs = db.get_table('graph.Edges_N_Organisation_N_FundingRound', limit=limit)
frs_orgs = pd.DataFrame(frs_orgs, columns=['org_id', 'fr_id', 'action', 'is_lead_investor', None, None]).drop(columns=[None])

frs_orgs.head()
```

```python
# Association: Funding rounds <-> People
frs_people = db.get_table('graph.Edges_N_Person_N_FundingRound', limit=limit)
frs_people = pd.DataFrame(frs_people, columns=['person_id', 'fr_id', 'action', 'is_lead_investor', None, None]).drop(columns=[None])

frs_people.head()
```

```python
# Association: Concepts <-> Organisations
concepts_orgs = db.get_table('graph.Edges_N_Organisation_N_Concept', limit=limit)
concepts_orgs = pd.DataFrame(concepts_orgs, columns=['org_id', 'concept_id', None, None]).drop(columns=[None])

concepts_orgs.head()
```

```python
# Association: EPFL Startups <-> Organisations
startups_orgs = db.get_table('graph.Edges_N_EPFLStartup_N_Organisation', limit=limit)
startups_orgs = pd.DataFrame(startups_orgs, columns=['startup_id', 'org_id', None, None]).drop(columns=[None])

startups_orgs.head()
```

```python
# Extract lists of ids of all models with at least one association
org_ids = list(set(frs_orgs['org_id']) | set(concepts_orgs['org_id']) | set(startups_orgs['org_id']))
fr_ids = list(set(frs_orgs['fr_id']) | set(frs_people['fr_id']))
person_ids = list(set(frs_people['person_id']))
concept_ids = list(set(concepts_orgs['concept_id']))
startup_ids = list(set(startups_orgs['startup_id']))
```

```python
# Filter models with no associations
orgs = orgs[orgs['org_id'].isin(org_ids)]
frs = frs[frs['fr_id'].isin(fr_ids)]
people = people[people['person_id'].isin(person_ids)]
concepts = concepts[concepts['concept_id'].isin(concept_ids)]
startups = startups[startups['startup_id'].isin(startup_ids)]
```

```python
len(org_ids), len(fr_ids), len(person_ids), len(concept_ids), len(startup_ids)
```

```python
# Unify missing values and format of some fields
orgs.loc[orgs['n_employees'] == 'unknown', 'n_employees'] = None
people.loc[people['gender'] == 'not_provided', 'gender'] = None

startups['industry'] = startups['industry'].str.lower()
```

```python
orgs
```

```python
orgs = orgs.merge(startups_orgs, how='left', on='org_id').merge(startups, how='left', on='startup_id')
```

```python
orgs = orgs[['org_id', 'org_name']]
frs = frs[['fr_id', 'fr_name', 'fr_date', 'fr_amount_usd']]
people = people[['person_id', 'full_name']]
```

```python
investments = concepts.merge(concepts_orgs, how='inner', on='concept_id')
investments = investments.merge(orgs, how='inner', on='org_id')
investments = investments.merge(frs_orgs[frs_orgs['action'] == 'Raised from'], how='inner', on='org_id')
investments = investments.merge(frs, how='inner', on='fr_id')

investments['year'] = investments['fr_date'].astype(str).str.split('-').str[0].astype(int)
```

```python
min_year = min(investments['year'])
max_year = max(investments['year'])

skeleton = pd.DataFrame({'concept_id': list(set(investments['concept_id']))})
skeleton = skeleton.merge(pd.DataFrame({'year': range(min_year, max_year + 1)}), how='cross')

skeleton
```

```python
cols = ['concept_id', 'concept_name', 'year']
investments = investments[cols + ['fr_amount_usd']].groupby(cols, as_index=False).sum()
```

```python
n_top_concepts = 10

top_concepts = investments[['concept_id', 'concept_name', 'fr_amount_usd']].groupby(['concept_id', 'concept_name'], as_index=False).sum()
top_concepts = top_concepts.sort_values(by='fr_amount_usd', ascending=False)
top_concepts = top_concepts[:n_top_concepts]

top_concept_ids = list(top_concepts['concept_id'])
```

```python
top_concept_investments = investments[investments['concept_id'].isin(top_concept_ids)]
```

```python
top_concept_investments = top_concept_investments.pivot_table(values=['fr_amount_usd'], columns=['concept_name'], index=['year'], aggfunc=sum)
top_concept_investments = top_concept_investments.fillna(0)
```

```python
fig, ax = plt.subplots(dpi=150)

from_years_ago = 10
to_years_ago = 1

ax.plot(top_concept_investments[-from_years_ago:][:-to_years_ago], label=top_concepts['concept_name'])
ax.legend()
```
